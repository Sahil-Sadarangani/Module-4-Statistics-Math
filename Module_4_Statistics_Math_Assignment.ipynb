## Question 2: Delivery Time Variability Analysis (Range, Variance, Standard Deviation)


```python
delivery_times = np.array([
3,5,2,4,6,2,3,4,2,5,7,2,3,4,2,4,2,3,5,6,
3,2,1,4,2,4,5,3,2,7,2,3,4,5,1,6,2,4,
3,5,3,2,4,2,6,3,2,4,5,3
])

range_val = np.max(delivery_times) - np.min(delivery_times)
variance_val = np.var(delivery_times)
std_dev_val = np.std(delivery_times)

print("Range of delivery times is:", range_val)
print("Variance of delivery times is:", variance_val)
print("Standard deviation of delivery times is:", std_dev_val)
```

## Question 4: Vehicle Fuel Efficiency Analysis (Mean, Range, Variance)




```python
models = {
    "Model A": np.array([30,32,33,28,31,30,29,30,32,31]),
    "Model B": np.array([25,27,26,23,28,24,26,25,27,28]),
    "Model C": np.array([22,23,20,25,21,24,23,22,25,24]),
    "Model D": np.array([18,17,19,20,21,18,19,17,20,19]),
    "Model E": np.array([35,36,34,35,33,34,32,33,36,34])
}

for model, values in models.items():
    mean_val = np.mean(values)
    range_val = np.max(values) - np.min(values)
    var_val = np.var(values)
    
    print(model, "- Mean:", mean_val, ", Range:", range_val, ", Variance:", var_val)
```

### Explanation:
Mean compares fuel efficiency, while range and variance show consistency across vehicle models.


## Question 5: Manufacturing Defect Frequency Analysis (Bar Chart and Histogram)

## Question 6: Customer Satisfaction Ratings Analysis (Skewness and Kurtosis)

## Question 11: Confidence Interval for Population Mean Height


```python
mean = 170
std = 8
n = 100
z = 1.96

margin_error = z * (std / np.sqrt(n))
lower = mean - margin_error
upper = mean + margin_error

print("95% Confidence Interval is:", (lower, upper))
```

### Explanation:
Confidence interval estimates the range in which the true population mean lies.

## Question 14: Confidence Interval for Population Proportion (Policy Support)


```python
n = 500
x = 320
p_hat = x / n
z = 1.645

margin = z * np.sqrt((p_hat * (1 - p_hat)) / n)
lower = p_hat - margin
upper = p_hat + margin

print("90% Confidence Interval for population proportion is:", (lower, upper))
```

### Explanation:
Proportion confidence interval estimates public support level with uncertainty bounds.
